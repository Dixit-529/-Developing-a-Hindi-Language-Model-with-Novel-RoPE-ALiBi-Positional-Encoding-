```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import lxml
import re
import pickle
PATH = 'D:/jupyter notebook works/wikipedia_hindi_articles_data/'
mapping_dict = {}
```

```python
# Reading the all links from the file
with open('D:/jupyter notebook works/all_hindi_wikipedia_links.pkl', 'rb') as f:
    all_urls = pickle.load(f)
```

```python
x = 214452
```

```python
# web scrapping all the links and get hindi data 
# prepocessing the data and then saved it in the pickle files
for counter, url in enumerate(all_urls[x:]):
    counter+=x
    html_doc = ''
    try:
        with urlopen(url) as response:
            for line in response:
                line = line.decode('utf-8')
                html_doc = html_doc + line.replace('\n','')
        soup = BeautifulSoup(html_doc, 'lxml')
        print(soup.title.string)
        paras = soup.find_all('p')
        article = ''
        for para in paras:
            article = article + para.text + '\n'
        article = re.sub(r'\([^)]*\)', r'', article)
        article = re.sub(r'\[[^\]]*\]', r'', article)
        article = re.sub(r'<[^>]*>', r'', article)
        article = re.sub(r'^https?:\/\/.*[\r\n]*', '', article)
        article = article.replace(u'\ufeff','')
        article = article.replace(u'\xa0', u' ')
        article = article.replace('  ', ' ')
        article = article.replace(' , ', ', ')
        devanagari_nums = ('०','१','२','३','४','५','६','७','८','९')
        for c, n in enumerate(devanagari_nums):
            article = re.sub(n, str(c), article)
            article = re.sub(n, c , article)
        filename = PATH + "article_"+str(counter) + '.pkl'
        mapping_dict[filename] = soup.title.string
        with open(filename, 'wb',encoding='utf-8') as f:
            pickle.dump(article,f)
        print("Saved " + filename)
    except:
        continue
```

# By multithreading

```python
import threading
from concurrent.futures import ThreadPoolExecutor
```

```python
urls = [[i,all_urls[i]] for i in range(216689,len(all_urls))]
```

```python
def save_file(x):
    counter = x[0]
    url = x[1]
    html_doc = ''
    try:
        with urlopen(url) as response:
            for line in response:
                line = line.decode('utf-8')
                html_doc = html_doc + line.replace('\n','')
        soup = BeautifulSoup(html_doc, 'lxml')
        print(soup.title.string)
        paras = soup.find_all('p')
        article = ''
        for para in paras:
            article = article + para.text + '\n'
        article = re.sub(r'\([^)]*\)', r'', article)
        article = re.sub(r'\[[^\]]*\]', r'', article)
        article = re.sub(r'<[^>]*>', r'', article)
        article = re.sub(r'^https?:\/\/.*[\r\n]*', '', article)
        article = article.replace(u'\ufeff','')
        article = article.replace(u'\xa0', u' ')
        article = article.replace('  ', ' ');
        article = article.replace(' , ', ', ');
        devanagari_nums = ('०','१','२','३','४','५','६','७','८','९')
        for c, n in enumerate(devanagari_nums):
            article = re.sub(n, str(c), article)
            article = re.sub(n, c, article)
        filename = PATH + "article_"+str(counter) + '.pkl'
        mapping_dict[filename] = soup.title.string
        with open(filename, 'wb') as f:
            pickle.dump(article,f)
        print("Saved " + filename)
    except:
        pass
```

```python
demo = urls
```

```python
threads = []
for x in demo:
    thread = threading.Thread(target=save_file,args=(x,)) 
    threads.append(thread)
    thread.start()
    
for thread in threads:
    thread.join()
```
