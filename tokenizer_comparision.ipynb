```python
import os
from glob import glob
from transformers import AutoTokenizer
import pandas as pd
import matplotlib.pyplot as plt
import time

# Configuration
TOKENIZERS = {
    "My Tokenizer": "/content/drive/MyDrive/my_hindi_t5_tokenizer",
    "IndicBERT": "ai4bharat/indic-bert",
    "MuRIL": "google/muril-base-cased",
    "XLM-R": "xlm-roberta-base"
}

TEXT_FILES_DIR = "/content/drive/MyDrive/gayatrivenugopal Hindi-Aesthetics-Corpus master Corpus/"
RESULTS_FILE = "/content/drive/MyDrive/tokenizer_comparison.csv"
LOG_FILE = "/content/drive/MyDrive/detailed_metrics.log"

def load_text_files(directory):
    """Load all text files from a directory"""
    files = glob(os.path.join(directory, "*.txt"))
    contents = []
    count = 0
    for file_path in files:
        if count>=1:
            break
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                contents.append(f.read())
        except Exception as e:
            print(f"Error loading {file_path}: {str(e)}")
        count+=1

    return contents

def analyze_tokenizer(tokenizer, texts):
    """Analyze a tokenizer across multiple texts"""
    metrics = {
        "total_tokens": 0,
        "total_oov": 0,
        "total_words": 0,
        "total_subwords": 0,
        "total_time": 0,
        "total_chars": 0
    }

    for text in texts:
        # Preprocess text
        words = text.strip().split()
        if not words:
            continue

        # Tokenization speed benchmark
        start_time = time.time()
        tokens = tokenizer.tokenize(text)
        metrics["total_time"] += time.time() - start_time

        # Calculate metrics
        metrics["total_tokens"] += len(tokens)
        metrics["total_oov"] += sum(1 for t in tokens if t not in tokenizer.get_vocab())
        metrics["total_words"] += len(words)
        metrics["total_subwords"] += sum(len(tokenizer.tokenize(word)) for word in words)
        metrics["total_chars"] += len(text)

    return {
        "vocab_size": tokenizer.vocab_size,
        "avg_oov_rate": (metrics["total_oov"] / metrics["total_tokens"]) * 100 if metrics["total_tokens"] else 0,
        "avg_tokens_per_word": metrics["total_subwords"] / metrics["total_words"] if metrics["total_words"] else 0,
        "avg_speed": metrics["total_tokens"] / metrics["total_time"] if metrics["total_time"] else 0,
        "compression_ratio": metrics["total_chars"] / metrics["total_tokens"] if metrics["total_tokens"] else 0,
        "special_tokens": len(tokenizer.special_tokens_map)
    }

def main():
    # Load all text files
    texts = load_text_files(TEXT_FILES_DIR)
    print(f"Loaded {len(texts)} text files for evaluation")

    # Initialize results storage
    results = []

    # Evaluate each tokenizer
    for name, model_path in TOKENIZERS.items():
        try:
            print(f"\nEvaluating {name}...")
            tokenizer = AutoTokenizer.from_pretrained(model_path)
            metrics = analyze_tokenizer(tokenizer, texts)
            results.append({"Tokenizer": name, **metrics})
        except Exception as e:
            print(f"Failed to evaluate {name}: {str(e)}")

    # Create DataFrame and save results
    df = pd.DataFrame(results)
    df.to_csv(RESULTS_FILE, index=False)
    print(f"\nSaved results to {RESULTS_FILE}")

    # Generate report
    with open(LOG_FILE, "w") as f:
        f.write("Tokenizer Comparison Report\n")
        f.write("="*30 + "\n")
        f.write(df.to_markdown(index=False))

    # Plot results
    metrics_to_plot = df.columns[1:]
    plt.figure(figsize=(15, 20))

    for i, metric in enumerate(metrics_to_plot, 1):
        plt.subplot(4, 2, i)
        plt.barh(df["Tokenizer"], df[metric])
        plt.title(metric.replace("_", " ").title())
        plt.xlabel(metric)

    plt.tight_layout()
    plt.savefig("tokenizer_comparison.png")
    print("\nGenerated visualization: tokenizer_comparison.png")

    # Show summary
    print("\nSummary Results:")
    print(df.to_markdown(index=False))

if __name__ == "__main__":
    main()
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoTokenizer
import sentencepiece as spm

# Configuration
RESULTS_CSV = "/content/drive/MyDrive/tokenizer_comparison_new.csv"
GRAPH_PATH = "/content/drive/MyDrive/tokenizer_comparison_new.png"

# Load your custom SentencePiece tokenizer
custom_tokenizer = spm.SentencePieceProcessor()
custom_tokenizer.load("/content/drive/MyDrive/hindi_tokenizer_bpe.model")
TEST_DATA_PATH = "/content/drive/MyDrive/hindi_corpus_fixed.txt"

# List of tokenizers to compare (add/remove as needed)
TOKENIZERS = {
    "My Tokenizer": custom_tokenizer,
    "IndicBERT": AutoTokenizer.from_pretrained("ai4bharat/indic-bert"),
    "MuRIL": AutoTokenizer.from_pretrained("google/muril-base-cased"),
    "mBERT": AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
}

def calculate_oov_rate(tokenizer, texts, name):
    """Calculate Out-of-Vocabulary rate"""
    if isinstance(tokenizer, spm.SentencePieceProcessor):
        return 0  # SentencePiece doesn't have explicit OOV tokens

    oov_count = 0
    total_tokens = 0

    for text in texts:
        tokens = tokenizer.tokenize(text)
        total_tokens += len(tokens)
        if tokenizer.unk_token:
            oov_count += tokens.count(tokenizer.unk_token)

    return (oov_count / total_tokens) * 100 if total_tokens > 0 else 0

def benchmark_tokenizer(tokenizer, name, texts):
    """Run benchmarks for a tokenizer"""
    results = {}

    # Vocabulary Size
    if isinstance(tokenizer, spm.SentencePieceProcessor):
        results["vocab_size"] = tokenizer.get_piece_size()
    else:
        results["vocab_size"] = tokenizer.vocab_size

    # Speed Test
    start_time = time.time()
    for text in texts:
        tokenizer.tokenize(text)
    results["speed_tps"] = len(texts)/(time.time() - start_time)

    # OOV Rate
    results["oov_rate"] = calculate_oov_rate(tokenizer, texts, name)

    # Avg Tokens per Sentence
    lengths = [len(tokenizer.tokenize(text)) for text in texts]
    results["avg_tokens"] = sum(lengths)/len(lengths)

    return results

def plot_results(df):
    """Generate comparison plots"""
    plt.figure(figsize=(15, 10))

    metrics = ["vocab_size", "speed_tps", "oov_rate", "avg_tokens"]
    titles = ["Vocabulary Size", "Speed (tokens/sec)", "OOV Rate (%)", "Avg Tokens per Sentence"]

    for i, metric in enumerate(metrics, 1):
        plt.subplot(2, 2, i)
        sns.barplot(x="tokenizer", y=metric, data=df)
        plt.title(titles[i-1])
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig(GRAPH_PATH)
    plt.close()

def main():
    # Load test data
    with open(TEST_DATA_PATH, "r", encoding="utf-8") as f:
        texts = [line.strip() for line in f.readlines()][:1000]  # Use first 1000 lines for testing

    # Run benchmarks
    results = []
    for name, tokenizer in TOKENIZERS.items():
        print(f"Testing {name}...")
        metrics = benchmark_tokenizer(tokenizer, name, texts)
        metrics["tokenizer"] = name
        results.append(metrics)

    # Save results
    df = pd.DataFrame(results)
    df.to_csv(RESULTS_CSV, index=False)

    # Generate plots
    plot_results(df)
    print(f"Results saved to {RESULTS_CSV} and {GRAPH_PATH}")

if __name__ == "__main__":
    main()
```
